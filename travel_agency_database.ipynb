### Bir fonksiyon oluşturun

`city` ve `key` olmak üzere iki argüman alan ve şehir için karşılık gelen değeri döndüren `get_key_value_from_city` fonksiyonunu oluşturun.

__Örnek__: `get_key_value_from_city(city=paris_city, key="total_population")` 2102650 döndürecektir.

<u>Not:</u> Eğer bir anahtar (key) yoksa, fonksiyon `"_<key>_ not defined"` döndürmelidir.
    




```python
def get_key_value_from_city(city, key):
    if key in city:
        return city[key]
    else :
        return f"_{key}_ not defined"
```

### Değerleri okuma


Önceki `get_key_value_from_city` fonksiyonunuzu kullanarak iki argüman alan yeni bir fonksiyon oluşturun: `city_name` ve `key`.

Eğer şehir listede yoksa, fonksiyon `"*city* doesn't exist or is not covered"` döndürmelidir.

Fonksiyonunuzu çeşitli örneklerle test edin.

```python
# Görev 1'den gelen yardımcı fonksiyon (Görev 2'nin çalışması için gerekli)
def get_key_value_from_city(city, key):
    if key in city:
        return city[key]
    else:
        return f"_{key}_ not defined"

# Görev 2: Değer okuma fonksiyonu
def read_city_value(city_name, key):
    # cities listesi içindeki her bir şehri kontrol et
    for city in cities:
        if city["name"] == city_name:
            # Şehir bulunduysa Görev 1'deki fonksiyonu çağır
            return get_key_value_from_city(city, key)
            
    # Döngü biter ve şehir bulunamazsa bu mesajı döndür
    return f"*{city_name}* doesn't exist or is not covered"

# Test etmek için:
# print(read_city_value("Paris", "total_population"))
# print(read_city_value("Istanbul", "continent"))
```

### Değer güncelleme


Bazen şehir bilgileri yanlış olabilir veya güncellemeye ihtiyaç duyabilir.

`modify_city_key_value` fonksiyonunu oluşturun; bu fonksiyon mevcut bir anahtarın değerini yeni bir değerle değiştirir. \
Fonksiyon üç argüman alır: `city_name`, `key` ve anahtara atanacak yeni `value`.

Fonksiyon aşağıdaki durumlardan birini belirten bir metin mesajı üretecektir:

- Eğer şehir yoksa veya kapsanmıyorsa: `"*city* doesn't exist or is not covered."`
- Eğer verilen anahtar yoksa: `"*key* doesn't exist."`
- Eğer anahtar başarıyla güncellendiyse: `"*key* has been updated for *city*."`

```python
# Görev 2: Değer güncelleme fonksiyonu
def modify_city_key_value(city_name, key, value):
    for city in cities:
        if city["name"] == city_name:
            # Şehir bulundu, şimdi anahtarın (key) olup olmadığını kontrol edelim
            if key in city:
                # Anahtar varsa yeni değeri ata
                city[key] = value
                return f"*{key}* has been updated for *{city_name}*."
            else:
                # Şehir var ama anahtar yoksa
                return f"*{key}* doesn't exist."
                
    # Şehir hiç yoksa
    return f"*{city_name}* doesn't exist or is not covered."

# Test etmek için:
# print(modify_city_key_value("Brooklyn", "total_population", 3000000))
```
